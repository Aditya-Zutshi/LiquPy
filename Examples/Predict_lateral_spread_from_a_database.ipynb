```python
# import libraries
from liqupy.points import calc_ls_bartlett, calc_ls_bardet, calc_ls_javadi2006, calc_ls_javadi_moderate2006, calc_ls_rezania, calc_ls_baziar2013, calc_ls_goh  # import the lateral spread analysis at points from LiquPy library

from liqupy.points import verify_ls_model # import verification method from LiquPy library

from liqupy.default_datasets import load_youd_bartlett_demo # default demo dataset of LiquPy
```

```python
# load dataset from LiquPy's default datasets
# NOTE: you may load your own dataset as a Pandas DataFrame and use it as the input;
#       in this case you must make sure your input data has the same format as the table below
data = load_youd_bartlett_demo()

# have a peek at dataset
data.head()
```

```python
# calculation lateral spread magnitudes using Bartlett's model on the dataset
verify_ls_model(calc_ls_bartlett, data)
```

```python
# calculation lateral spread magnitudes using other models such as Bardet's model 
# on the dataset is as simple as changing the first argument 
verify_ls_model(calc_ls_bardet, data)
```
